```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/delay_data/delay_mpr_60_flow_54_5_time_300.csv')
```

```python
data['vehicle_type'] = data['vehicle_id'].apply(lambda x: 'HDV' if 'HDV' in x else 'CAV')
plt.figure(figsize=(8, 6))
sns.boxplot(x='vehicle_type', y='delay_time', data=data)
plt.title('Delay Time by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Delay Time')
plt.show()
```

```python
data = pd.read_csv('../data/delay_data/delay_mpr_60_flow_54_12_time_300_without_ramp.csv')
```

```python
data['vehicle_type'] = data['vehicle_id'].apply(lambda x: 'HDV' if 'HDV' in x else 'CAV')
plt.figure(figsize=(8, 6))

# 指定 HDV 在左，CAV 在右的顺序
sns.boxplot(x='vehicle_type', y='delay_time', data=data, order=['HDV', 'CAV'])

plt.title('Delay Time by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Delay Time')
plt.show()
```
